```python
#Session - 25
#LINEAR ALGEBRA FUNCTION FROM LINALG() MODULE  
#inv() function
import numpy as np
a=np.arange(8).reshape(2,2,2)
print(np.linalg.inv(a))

print()
#matrix_power()
a=np.array([[1,2],[3,4]])
print(np.linalg.matrix_power(a,0))
print(np.linalg.matrix_power(a,2))
print(np.linalg.matrix_power(a,-2))
print()

#det() function (determinant function)
print(np.linalg.det(a))
a=np.arange(8).reshape(2,2,2)
print(np.linalg.det(a))
print()

#solve() function
a=[[1,1],[3,8]]
b=np.array([2200,10100])
print(np.linalg.solve(a,b))

print()
a=np.array([[-4,7,-2],[1,-2,1],[2,-3,1]])
b=np.array([2,3,-4])
print(np.linalg.solve(a,b))


```
